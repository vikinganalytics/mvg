```python
from mvg import MVG
from mvg.analysis_classes import parse_results

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# env vars
from dotenv import load_dotenv
load_dotenv()
```

```python
# Replace token and endpoint with your token and server address respectively
ENDPOINT = os.environ["MVG_URL"]
TOKEN = os.environ["MVG_TOKEN"]
session = MVG(ENDPOINT, TOKEN)
print("Connected to MVG API")
```

```python
# Get ModeId results
request_id = "5851134365a0b7823580ff10f0c50385"
ana_response = session.get_analysis_results(request_id)
ana_results = parse_results(ana_response, t_zone="UTC", t_unit="ms")
```

```python
ana_results.plot_probabilities(selected_modes=[])
```
